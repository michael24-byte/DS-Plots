## A Distribution Plot that shows the various ways to display the distributions of a dataset.

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

#### Totalbills Distribution

```python
sns.displot(tips["total_bill"]);
```

### How Does Tips Compare with Totalbills

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips );
```

#### Make a PairPlot

```python
sns.pairplot(tips, hue= "sex");
```

#### Kde Plot

```python
sns.kdeplot(tips["total_bill"]);
```
