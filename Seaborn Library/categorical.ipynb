### Visualizing Categorical Data

```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.barplot(x = "sex", y = "total_bill", data = tips, hue= "sex", palette= "Set2", estimator= np.std);
```

```python
sns.countplot(x = "sex", data = tips, hue= "sex", palette= "Set2");
```

#### Totalbills Per Day

```python
sns.boxplot(x= "day", y= "total_bill", data= tips, hue = "day", palette= "Set3");
```

```python
sns.violinplot(x= "day", y= "total_bill", data= tips, hue = "sex", split= True, palette= "Set2");
```

```python
sns.stripplot(x= "day", y= "total_bill", data= tips, hue = "sex", jitter= True, palette= "Set2");
```

#### The more general method

```python
sns.catplot(x= "day", y= "total_bill", data= tips, kind= "bar", hue= "day" )
```
