### This displays some common datascience plots with a univariate.

Import Libraries

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np 
```

```python
#Generate Data
data = np.random.rand(1000)
```

```python
#Plot a histogram
sns.histplot(data, kde= True)
plt.title("Histogram with Kde")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

#Save the figure
plt.savefig("my_histplot.png", dpi=300)
```

#### Boxplot

```python
sns.boxplot(x = data)
plt.title("Boxplot")
plt.xlabel("value")
plt.show()
```

#### Boxplot: Group Data

```python
import pandas as pd
```

```python
#Create Droup Data
group_data = pd.DataFrame({
    "value": np.concatenate([np.random.normal(loc, 1, 100) for loc in [0, 1, 2]]),
    "group": ["A"]*100 + ["B"]*100 + ["C"]*100
})
```

```python
sns.boxplot(x="group", y="value", data=group_data)
plt.title("Group Boxplot")
plt.xlabel("Group")
plt.ylabel("Value")
plt.show()
```

#### Violin Plot

```python
sns.violinplot(x=data)
plt.title("Violin Plot")
plt.xlabel("Value")
plt.show()
```

#### BarChart

```python
# Sample categorical data
data1 = pd.Series(['A', 'B', 'A', 'C', 'B', 'A', 'C', 'C', 'C', 'B'])

sns.countplot(x = data1)
plt.title("Univariate Bar Chart - Category Counts")
plt.show()
```
